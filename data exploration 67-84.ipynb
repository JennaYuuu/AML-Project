```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset.csv')
df
```

```python
df_67_to_84 = df.iloc[:, 67:84]
df_67_to_84 = df_67_to_84.join(df['hospital_death'])
df_67_to_84
```

```python
df_67_to_84 = df_67_to_84.drop(columns=['Unnamed: 83'])
```

```python
df = df_67_to_84
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 6), sharey=True)
sns.boxplot(ax = axes[0], data=df, y='d1_glucose_max', x='hospital_death')
sns.boxplot(ax = axes[1], data=df, y='d1_glucose_min', x='hospital_death')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 6), sharey=True)
sns.boxplot(ax = axes[0], data=df, y='d1_potassium_max', x='hospital_death')
sns.boxplot(ax = axes[1], data=df, y='d1_potassium_min', x='hospital_death')
plt.show()
```
