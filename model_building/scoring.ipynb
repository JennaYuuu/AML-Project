```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns = ['Model', 'Accuracy', 'ROC_AUC', 'Average_Precision', 'F1']
```

```python
df_scoring = pd.DataFrame(columns = columns)
df_scoring = df_scoring.append(pd.DataFrame(data = [['Logistic', 0.924, 0.62, 0.366, 0.232]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['Neural Network', 0.9229, 0.8514, 0.2433, 0.4017]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['Random Forest', 0.9296, 0.6302, 0.2673, 0.3968]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['Histgradient Boosting', 0.9293, 0.6593, 0.2860, 0.4483]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['LinearSVC (L1)', 0.8016, 0.8736, 0.4691, 0.4005]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['LinearSVC (L1)', 0.8568, 0.8697, 0.4067, 0.4509]], columns=columns), ignore_index=True)
```

```python
df_scoring
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=2)
ax = sns.barplot(data=df_scoring, x="Model", y="Accuracy", hue = 'Model', dodge=False)
for i in ax.containers:
    ax.bar_label(i,)
ax.get_legend().remove()
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=2)
ax = sns.barplot(data=df_scoring, x="Model", y="Average_Precision", hue = 'Model', dodge=False)
for i in ax.containers:
    ax.bar_label(i,)
ax.get_legend().remove()
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=2)
ax = sns.barplot(data=df_scoring, x="Model", y="ROC_AUC", hue = 'Model', dodge=False)
for i in ax.containers:
    ax.bar_label(i,)
ax.get_legend().remove()
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=2)
ax = sns.barplot(data=df_scoring, x="Model", y="F1", hue = 'Model', dodge=False)
for i in ax.containers:
    ax.bar_label(i,)
ax.get_legend().remove()
```
