```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns = ['Model', 'Accuracy', 'ROC_AUC', 'Average_Precision', 'F1']
```

```python
df_scoring = pd.DataFrame(columns = columns)
df_scoring = df_scoring.append(pd.DataFrame(data = [['Logistic', 0.924, 0.62, 0.366, 0.232]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['Neural Network', 0.923, 0.851, 0.243, 0.402]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['Random Forest', 0.930, 0.630, 0.267, 0.397]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['Histgradient Boosting', 0.929, 0.659, 0.286, 0.448]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['LinearSVC (L1)', 0.802, 0.874, 0.469, 0.401]], columns=columns), ignore_index=True)
df_scoring = df_scoring.append(pd.DataFrame(data = [['SVC(RBF kernel)', 0.857, 0.870, 0.407, 0.451]], columns=columns), ignore_index=True)
```

```python
df_scoring = df_scoring.reset_index()
df_scoring = pd.melt(df_scoring, id_vars='Model', var_name ='Metrics', value_vars=['Accuracy', 'ROC_AUC', 'Average_Precision', 'F1'])
```

```python
df_scoring
```

```python
plt.figure(figsize=(20,10))
sns.set(font_scale=1.5)
ax = sns.barplot(data=df_scoring, x='Metrics', y='value', hue='Model')
for i in ax.containers:
    ax.bar_label(i,rotation=30)
```
