```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from google.colab import files
import io
from matplotlib import colors
from matplotlib.ticker import PercentFormatter
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/tzhou19/AML_project/main/Data/dataset.csv')
```

```python
feature_names=df.columns
perc=df.isnull().sum()
percx=[]
for element in perc:
  percx.append((element/df.shape[0])*100)

fig= plt.figure(figsize=(16,10))
axs=sns.barplot(x=feature_names,y=percx)
axs.tick_params(axis='x',rotation=90)
plt.xlabel("Columns")
plt.ylabel("% missing")
plt.show()
```

```python
def hist_feat():
  fig, axs = plt.subplots(1, 1,figsize =(16, 10),tight_layout = True)
  
 
  # Remove axes splines
  for s in ['top', 'right']:
      axs.spines[s].set_visible(False) 
    
  # Add padding between axes and labels
  axs.xaxis.set_tick_params(pad = 5)
  axs.yaxis.set_tick_params(pad = 10)
  
  # Add x, y gridlines
  axs.grid(b = True, color ='grey', linestyle ='-.', linewidth = 0.5, alpha = 0.6)
```

```python
for name in feature_names:
  hist_feat()
  sns.histplot(df[name])
  plt.show()  
```
