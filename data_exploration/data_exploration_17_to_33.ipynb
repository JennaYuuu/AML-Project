```python
corr_matrix = df_17_to_33.corr()
plt.figure(figsize=(20,12))
sns.heatmap(corr_matrix, cmap="Greens", annot=True)
```

```python
percent_null = df_17_to_33.isnull().sum() * 100 / len(df_17_to_33)
ax = sns.barplot(y=df_17_to_33.columns, x=percent_null.values)
ax.set(ylabel='features', xlabel='percent of missing values')
sns.set(rc={'figure.figsize': (20, 10)})
plt.show()
```
